**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collections and Processing**

```python
data = pd.read_csv('/content/train (2).csv')
data
```

```python
# printing the first 5 rows of the dataset
```

```python
# Getting the Info about the data
data.info()
```

```python
# No of missing values in each column
```

```python
data.isnull().sum()
```

```python
# Handling the missing values
# Dropping the "Cabin" columns from the dataframe
data = data.drop(columns='Cabin',axis=1)
```

```python
data.head()
```

```python
# reolacing the missing values in the "age" column with the mean values
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
```

```python
data.isnull().sum(  )
```

```python
# finding the mode value of the embarked column
data['Embarked'].value_counts()
```

```python
print(data['Embarked'].mode())
```

```python
# replacing the missing values in the "Embarked" column with mode value
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
# checking the number of mkising values again
data.isnull().sum()  # no missing values
```

**Data Analysis**

```python
# Getting the Statistical Details of the Dataset
data.describe()
```

```python
# Finding the number of people survived and not survived
data['Survived'].value_counts()
```

```python
# Finding the number of Male and Female passengers
data['Sex'].value_counts()
```

```python
# Lets check the number if people class wise
data['Pclass'].value_counts()
```

**Encoding the Categorical Column**

```python
data['Sex'].value_counts()
```

```python
data['Embarked'].value_counts()
```

```python
data = data.replace({'Sex':{'male':1,'female':0},'Embarked':{'S':0,'C':1,'Q':2}})
```

```python
data.head()
```

**Seperating the Features and the Target**

```python
X = data.drop(columns =['PassengerId','Name','Ticket','Survived'],axis=1)
```

```python
y = data.Survived
```

```python
X
```

```python
y
```

**Splitting the Data into Traning and Testing**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# **Model Training**

**Logistic Regression**

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with the training data
```

```python
model.fit(X_train,y_train)
```

Model Evaluation

```python
# Accuracy Score
```

```python
# accuracy_score on training data
```

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(y_train,X_train_prediction)
print("Accuracy Score of training data",training_data_accuracy)
```

```python
# Accuracy on the Testing Data
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score = accuracy_score(y_test,y_pred)
print ("Accuracy_score is:",accuracy_score)
```
